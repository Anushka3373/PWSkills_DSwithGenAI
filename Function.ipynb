7. What are the advantages of using generators over regular functions?
   - 1> memory efficiency - Generators do not store the entire sequence in memory.
   - 2> Lazy evaluation - values are produed only when needed , not all at once.
   - 3> Infinite sequence - You can create sequence that never end(like a number stream).

```python
def count(n):
  for i in range(n):
    yield i
gen = count(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

   - **map()**
   - Applies the function to each item of an iterable and returns a new iterable.

```python
nums = [1, 2, 3]
result = map(lambda x: x*2, nums)
print(list(result))
```

   - **reduce()**
   - Applies a function to accumulate a result, reduing the iterable to single value.
   

```python
from functools import reduce
nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x+y, nums)
print(result)
```

   - **filter()**
   - Applies a function that returns true/false to each items and returns only the items that are true.
   

```python
nums = [1, 2, 3, 4]
result = filter(lambda x: x % 2 == 0, nums)
print(list(result))
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum(numbers):
  even_sum=0
  for num in numbers:
    if num % 2 == 0:
      even_sum += num
  return even_sum
lst = [1, 2, 3, 4, 5, 6]
result = sum(lst)
print("sum of even numbers:",result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
  if n <= 1:
    return False
  for i in range(2, n):
    if n % i == 0:
      return False
  return True
for num in range(1, 201):
  if is_prime(num):
    print(num,"is a prime number")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.first = 0
        self.second = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.first
        elif self.count == 1:
            self.count += 1
            return self.second
        else:
            self.first, self.second = self.second, self.first + self.second
            self.count += 1
            return self.second
```

```python
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(max_exponent):
  for exp in range(max_exponent + 1):
    yield 2**exp
for value in power_of_two(5):
  print(value)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
with open("Sample.txt", "w") as file:
  file.write("Line 1: Hello World!\n")
  file.write("Line 2: This is a text file\n")
  file.write("Line 3: I am Anushka")
```

```python
def read_file(filename):
  with open(filename, 'r') as file:
    for line in file:
      yield line.strip()
for line in read_file("Sample.txt"):
  print(line)
```
